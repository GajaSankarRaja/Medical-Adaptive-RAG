```python
import matplotlib.pyplot as plt
import numpy as np

# Data
sizes = np.array([48,22,30])
labels = ["Vector-Based", "Knowledge-Graph", "Hybrid(Vector + Graph)"]
explode = [0, 0, 0] # Pull the 'Apples' slice out
colors = ['yellowgreen', 'lightcoral', 'lightskyblue'] # Custom colors

# Create the customized pie chart
plt.pie(sizes,
        labels=labels,
        explode=explode,
        colors=colors,
        autopct='%1.1f%%', # Display percentages
        shadow=False,
        startangle=44)

# Ensure the chart is a perfect circle
plt.axis('equal')

# Add a title
plt.title('Result Strategy selection')
#plt.legend(loc='lower right')
# Display the chart
plt.show()
```
